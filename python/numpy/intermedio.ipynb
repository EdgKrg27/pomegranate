# Intermedio -> Manipulación, Indexación y Operaciones

```python
import numpy as np
```

```python
# Crea una matriz 2D y haz una transposición
arr = np.random.randint(40, 100, size=(2, 2))
print(arr)
print()
print(arr.transpose())
```

```python
# Considera un array A de forma (5,5,3). Calcula la suma a lo largo del ultimo eje
arr = np.random.rand(5, 5, 3)
print(f"Array de forma {arr.shape}")
suma = np.sum(arr, axis=2)
print(f"Suma a lo largo del ultimo eje {suma.shape}")
print(suma)
```

```python
# Intercambia la primera y la última fila de una matriz 5x5 dada
random_matrix = np.random.randint(1, 100, size=(5, 5))
print(random_matrix, "\n")
random_matrix[[0, -1]] = random_matrix[[-1, 0]]
print(random_matrix)
```

```python
# Crea un array de números aleatorios y encuentra los 5 valores más grandes
arr = np.random.randint(0, 100, size=(15))
print(arr)
k = 5
max_value = np.partition(arr, -k)[-k:]
print(max_value)
```

```python
# Con una matriz 2D, cálcula la media a lo largo de las columnas
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)
res = np.mean(arr, axis=0)
print(res)
```

```python
# Dada una matriz 2D, selecciona solo las filas donde el valor de la primera columna sea mayor que 5
arr = np.random.randint(0, 10, size=(5, 7))
print(arr)
res = arr[arr[:, 0] > 5]
print("\n", res)
```

```python
# ¿Cómo se puede hacer broadcasting para sumar un array 1D a cada fila de un array 2D?
arr1 = np.random.randint(10, 60, size=(5))
print(arr1)
arr2 = np.random.randint(10, 60, size=(2, 5))
print(arr2)
res = arr1 + arr2
print(res)
```

```python
# Crea un array 1D y dividelo (split) en 3 sub-arrays
arr = np.random.randint(0, 10, size=(12))
print(arr)
print(arr[:4])
print(arr[4:8])
print(arr[8:])
```

```python
# Considera dos arrays A y B. Determina qué elementos de A también está presentes en B
A = np.array([1, 2, 3, 4, 5, 6])
B = np.array([4, 5, 6, 7, 8, 9])
inter = np.intersect1d(A, B)
print(inter)
```

```python
# Cálcula la desviación estándar de una matriz 3x3 de números aleatorios
arr = np.random.rand(10)
print(arr)
res = np.std(arr)
print(res)
```

```python
# ¿Cómo se elimina las filas que contienen valores NaN de una matriz 2D?
matrix = np.array(
    [[1, 2, 3], [4, np.nan, 6], [7, 8, 9], [np.nan, np.nan, np.nan], [10, 11, 12]]
)
print(matrix)
res = matrix[~np.any(np.isnan(matrix), axis=1)]
print(res)
```

```python
# Crea una matriz NxN, crea un patrón de tablero de ajedrez (alternando 0 y 1)
table = np.zeros((8, 8), dtype=int)
table[1::2, ::2] = 1  # filas impares
table[::2, 1::2] = 1
print(table, "\n")
```

```python
# Ordena las filas de una matriz 2D basándote en los valores de la segunda columna
matrix = np.array([[5, 2, 8], [1, 9, 4], [3, 1, 7], [7, 5, 2], [2, 3, 6]])
print(matrix)
print()

indices = np.argsort(matrix[:, 1])
print(indices)
print()

matriz_res = matrix[indices]
print(matriz_res)
```

```python
# ¿Cómo se puede convertir un array 1D de forma N a una matriz 2D de forma (1, N) (agregar una dimensión con newaxis)
arr = np.random.randint(5, 75, size=(7))
print(arr)
print(arr.shape)

arr2 = arr[np.newaxis, :]
print(arr2)
print(arr2.shape)
```

```python
# Cálcula el cuartil superior (75 percentil) de un array de 100 números aleatorios
data = np.array([12, 15, 16, 18, 20, 12, 18])
data.sort()
q3 = np.quantile(data, 0.75)
print(q3)
```

```python
# Encuentra la intersección (elementos comunes) de dos arrays 1D
arr1 = np.random.randint(1, 10, size=(7))
print(arr1)
arr2 = np.random.randint(1, 10, size=(7))
print(arr2)
res = np.intersect1d(arr1, arr2)
print(res)
```
