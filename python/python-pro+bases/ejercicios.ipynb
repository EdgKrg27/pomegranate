# Ejercicos de Python avanzado + bases 

## Intermedio
Ejercicios que refuerzan la sintaxis, el manejo básico de errores y las compresiones de estructuras

Escribe una función que reciba una cadena y devuelva un diccionario con la frecuencia de cada carácter, ignorando mayúsculas/minúsculas.

Dada una lista de números, usa un set para eliminar los duplicados y devolver la lista ordenada de menor a mayor.

Crea una lista con los cuadrados de los números del 1 al 20, pero solo si el número es par.

A partir de dos listas (una de llaves y otra de valores), crea un diccionario. Si la cantidad de elementos difiere, ajusta la más larga.

Genera un conjunto con las longitudes de las palabras en una frase, pero solo si la longitud es mayor a 3.

Pide al usuario dos números e intenta dividirlos. Captura ZeroDivisionError y ValueError mostrando mensajes distintos.

Intenta convertir una lista de strings a enteros usando un bucle. Si un elemento no se puede convertir, registra el error pero continúa con los demás.

Crea una función que reciba un número variable de argumentos (*args) y devuelva la suma de todos los números que sean mayores a 10.

 Crea una función que reciba una cadena y devuelva otra cadena donde cada palabra esté capitalizada, usando map y lambda.

 Aplana una lista que contiene sub-listas (ej. [[1,2], [3,4]]) en una sola lista usando una sola línea.

 Invierte las llaves y valores de un diccionario. Asume que los valores son únicos.

 Implementa una función que lea un archivo de texto. Si el archivo no existe, captura FileNotFoundError y crea el archivo vacío automáticamente.

 Escribe una función lambda que tome dos argumentos y devuelva el mayor de ellos. Asígnala a una variable y úsala dentro de un filter para filtrar números mayores a un umbral.

Usa collections.Counter para encontrar la palabra más común en una lista de strings.

Dada una matriz (lista de listas), extrae todos los elementos de la columna 2 (índice 1) usando comprensión.
